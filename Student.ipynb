```python
import matplotlib.pyplot as plt

# Educational data
categories = ["Paid", "Nursery", "Higher", "Activities"]
count = [40, 60, 75, 30]  # Sample data, you can replace it with your own

# Create a bar graph
plt.bar(categories, count, color=['blue', 'green', 'red', 'orange'])

# Add labels and a title
plt.xlabel("Categories")
plt.ylabel("Number of Students")
plt.title("Educational Data Analysis")

# Show the bar graph
plt.show()
```
