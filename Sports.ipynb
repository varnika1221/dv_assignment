```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Draw': [1, 2, 3, 4, 5],
    'Away Win': [12, 15, 8, 10, 14],
    'Home Win': [10, 9, 12, 8, 11],
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Calculate average scores
df['TeamA_Avg'] = df['Away Win'].mean()
df['TeamB_Avg'] = df['Home Win'].mean()

# Plot the scores
plt.figure(figsize=(10, 6))
plt.plot(df['Draw'], df['Away Win'], marker='o', label='Away Win')
plt.plot(df['Draw'], df['Home Win'], marker='o', label='Home Win')
plt.axhline(df['TeamA_Avg'][0], color='blue', linestyle='--', label='Away Win')
plt.axhline(df['Home Win'][0], color='red', linestyle='--', label='Home Win')
plt.xlabel('Draw')
plt.ylabel('Score')
plt.title('Sports Data Analysis')
plt.legend()
plt.grid(True)
plt.show()

# Display the DataFrame
print(df)
```
