```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the health data from a CSV file
data = pd.read_csv('fetal_health.csv')  # Replace 'health_data.csv' with your data file

# View the first few rows of the dataset to understand its structure
print(data.head())

# Create a histogram of a specific health metric (e.g., 'Blood Pressure')
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="uterine_contractions", kde=True)
plt.xlabel("uterine_contractions")
plt.title("Distribution of uterine_contractions")
plt.show()
```
